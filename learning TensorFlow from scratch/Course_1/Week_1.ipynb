---

1.   **models.Sequential()**

        * The **models.Sequential()** is used to create a simple batch of models, and layers are placed in it that have an input and an output.
        
        How to import it is as follows.

                from tensorflow.keras import models


2.   **Dense()** 

        * Dense implements the operation: 
        
        output = activation(dot(input, kernel) + bias) 
        
        where activation is the element-wise activation function passed as the  activation argument, kernel is a weights matrix created by the layer, and bias is a bias vector created by the layer (only applicable if use_bias is True). These are all attributes of Dense. 

        How to import it is as follows.

                from tensorflow.keras import models

3.   **tanh**

    * tanh(x) = sinh(x)/cosh(x) = ((exp(x) - exp(-x))/(exp(x) + exp(-x)))
    
    * Activations can either be used through an Activation layer, or through the activation argument supported by all forward layers:

    * Dense(64, activation=tf.nn.tanh)

4.   **sigmoid**

    * sigmoid(x) = 1 / (1 + exp(-x))
    
    * Activations can either be used through an Activation layer, or through the activation argument supported by all forward layers:

    * Dense(64, activation=tf.nn.sigmoid)

5.   **mean_squared_error**

    * Computes the mean of squares of errors between labels and predictions.This function is used for regression.


6.   **Adam**
    * Adam optimization is a stochastic gradient descent method that is based on adaptive estimation of first-order and second-order moments.

    * According to [ Kingma et al., 2014](http://arxiv.org/abs/1412.6980), the method is "computationally efficient, has little memory requirement, invariant to diagonal rescaling of gradients, and is well suited for problems that are large in terms of data/parameters".

    * You can go [here](https://www.youtube.com/watch?v=JXQT_vxqwIs) here for more information.


7.   **Usage with compile() & fit()**

    * In order to train our tensorflow model, we finally need to integrate these two commands.

    * However, the Four Ring can also be used for most models, both of which can be seen [here](https://keras.io/api/optimizers/) in more detail.










### example_1

```python
from tensorflow.keras import models
from tensorflow.keras.layers import Dense
import numpy as np
```

```python
X = np.array([[1,2],[3,4],[5,6],[7,8]])
y = np.array([5,7,9,11])

model = models.Sequential()
model.add(Dense(units=10,activation='tanh',input_shape = (None,2)))
model.add(Dense(units=1,activation='sigmoid'))
model.compile(optimizer = 'Adam', loss = 'mean_squared_error')
model.fit(X, y ,epochs=100)
```
